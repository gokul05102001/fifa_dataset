```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("fifa_data.csv")
```

```python
df.head()
```

# 1.Which country has the most number of players (score :1)

```python
country_count=df["Nationality"].value_counts()
```

```python
most_players_country=country_count.idxmax()
```

```python
print(f"country has the most number of players is :{most_players_country} -{country_count.max()} players")
```

# 2.Plot a bar chart of 5 top countries with the most number of players. (score :1)

```python
countries_top_5=country_count.head(5)
```

```python

plt.figure(figsize=(8, 6))
countries_top_5.plot(kind='bar', color='skyblue')
plt.title('Top 5 Countries with the Most Number of Players')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.show()
```

# 3.Which player has the highest salary? (score :1)

```python
df['Wage']=df['Wage'].str.replace('K','')
df['Wage']=df['Wage'].str.replace('€','')
df['Wage']=df['Wage'].astype(int)
df['Wage'].dtypes
```

```python
highest_salary=df.sort_values(by="Wage",ascending=False).head(1)
print(f"The Player with highest Salary is{highest_salary['Name']}")
```

# 4.Plot a histogram to get the salary range of the players. (score :1)

```python
plt.figure(figsize=(10,6))
plt.hist(df['Wage'],color='skyblue', edgecolor='black',orientation="horizontal",label="Number of Players")
plt.title('salary range of the Players')  # Add a title
plt.ylabel('Salary Range in k')  # Add x-axis label
plt.xlabel('Count')  # Add y-axis label
plt.xticks(range(0,19000,1000))
plt.grid(True)
plt.yticks(range(0,600,100))
plt.legend()
plt.show
```

# 5.Who is the tallest player in the fifa? (score :1)

```python
df['Height']=df['Height'].str.replace('\'','.')
df['Height']=df['Height'].astype(float)
df['Height'].dtypes
```

```python
tallest_player=df.sort_values(by="Height",ascending=False).head(1)
print(f"The Tallest Player is {tallest_player['Name']}")
```

# 7.Which foot is most preferred by the players?Draw a bar chart for preferred foot (score :1)

```python
df.value_counts(["Preferred Foot"])
```

```python
x=df.value_counts(["Preferred Foot"]).index
y=df.value_counts(["Preferred Foot"]).values
x1=list(x[0])
x2=list(x[1])
x=x1+x2
plt.figure(figsize=(10,6),dpi=100)
plt.bar(x,y,label="Number of Players")
plt.title('most prefered foot by the players')
plt.yticks(range(0,15000,1000))# y axis numbers
plt.xlabel('prefered foot')
plt.ylabel('Number of Players')
plt.legend() # shows label
plt.show()
```
